```python
import pandas as pd
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
csv_path = "Resources/gunviol.csv"
gunviol_df = pd.read_csv(csv_path)

gunviol_df.head()
```

```python
gunviol_df['day'] = ''
gunviol_df['month'] = ''
gunviol_df['year'] = ''
gunviol_df[["year", "month", "day"]] = gunviol_df["date"].str.split("-", expand = True)
gunviol_df
```

```python
list(gunviol_df.columns)
```

```python
newgunviol_df = gunviol_df[['state',
                            'n_killed',
                            'n_injured',
                            'year']]
newgunviol_df
```

```python
new_df= newgunviol_df[(newgunviol_df['year'] == '2017')]
new_df
```
